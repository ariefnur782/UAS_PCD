```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
from matplotlib import rcParams

img = cv2.imread("/kaggle/input/uas-pcdd/download.jpg")

k1 = np.ones((3,3),np.float)/9
k2 = np.ones((4,4),np.float)/16
k3 = np.ones((5,5),np.float)/25
k4 = np.ones((9,9),np.float)/81

img1 = cv2.filter2D(img,-1,k1)
img2 = cv2.filter2D(img,-1,k2)
img3 = cv2.filter2D(img,-1,k3)
img4 = cv2.filter2D(img,-1,k4)

rcParams['figure.figsize'] = 20,8
plt.subplot(1,5,1), plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)), plt.title("original")
plt.subplot(1,5,2), plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)), plt.title("3*3")
plt.subplot(1,5,3), plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)), plt.title("4*4")
plt.subplot(1,5,4), plt.imshow(cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)), plt.title("5*5")
plt.subplot(1,5,5), plt.imshow(cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)), plt.title("9*9")
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread("/kaggle/input/uas-pcd/pti (2).jpg")
kernel = np.ones((3, 3), np.float32)/9

img_out = cv.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```

```python
# Load image
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load image as grayscale
image = cv2.imread('/kaggle/input/uas-pcd/pti (2).jpg', cv2.IMREAD_GRAYSCALE)

# Create kernel
kernel = np.array([[0, -1, 0], 
                   [-1, 5,-1], 
                   [0, -1, 0]])

# Sharpen image
image_sharp = cv2.filter2D(image, -1, kernel)

# Show image
plt.imshow(image_sharp, cmap='gray'), plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt

img = cv2.imread("/kaggle/input/uas-pcd/pti (2).jpg", 0)
canny = cv2.Canny(img, 100, 200)

titles = ['image', 'canny']
images = [img, canny]
for i in range(2):
    plt.subplot(1, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()
```
